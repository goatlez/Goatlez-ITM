```python
from urllib.request import urlopen

url = "https://www.gutenberg.org/cache/epub/10/pg10.txt"

page = urlopen(url)

html_bytes = page.read()
html = html_bytes.decode("utf-8")

body_index = html.find("<body>")

start_index = body_index + len("<body>")

end_index = html.find("</body>")

pg10 = (html[start_index:end_index]).lower()

###################################

line_count = len(pg10.splitlines())

print("line count: ", line_count+1) 

#99969

###################################

content = pg10.split()

filter_indicator = 0
filtered_content = []
alphabet = [    '.', ',', ';', ':', '?', '!', '-', '(', ')', '[', ']', '"', "'", '&', '$', '+', '=', '©', '®', '¶'
,'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']


for word in content:
    for letter in word:
        if letter in alphabet:
            filter_indicator = filter_indicator + 0
        else: 
            filter_indicator = filter_indicator + 1
    if filter_indicator == 0:
        filtered_content.append(word)
    filter_indicator = 0

word_count = len(filtered_content)

print("word count: " , word_count)

#OLD: 824537

#NEW: 791271

##################################

word_counter = 0

for word in content:
    if word.lower() == "apostle" and "apostle!" and "apostle." and "apostle," and "apostle:" and "apostle's" and "apostles" :
        word_counter = word_counter + 1

print (f"number of occurences of the word 'apostle': {word_counter}")

# Number of occurences: 41 

#apostle_count = pg10.count('apostle')
 
#print ("'apostle' count: " , apostle_count)

# Number of occurences: 113

##################################


frequency_list = []
frequency_counter = 0

unique_list = list(set(filtered_content))

for word in unique_list:
    for word_variable in filtered_content:
        if word == word_variable:
            frequency_counter = frequency_counter + 1
    frequency_list.append(frequency_counter)
    frequency_counter = 0

max_value = max(frequency_list)

max_index = frequency_list.index(max_value)

print ("most frequent word: " , unique_list[max_index])
```
